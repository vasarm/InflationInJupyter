# Funktsioonidega arvutamine ning tulemuste kuvamine

Arvutusi on võimalik teha kahel viisil:
* 1\. Defineerida skalaarväljast sõltuvad funktsioonid $A(\phi)$, $B(\phi)$ ja $V(\phi)$,
* 2\. Defineerida invariantsest skalaarväljast sõltuv funktsioon $I_V(I_\phi)$ (invariantne potentsiaal)

## 1. Programmi kasutamine $A(\phi)$, $B(\phi)$ ja $V(\phi)$ korral

* 1\.1\. Impordime programmi:
```Python
    import modules as mod
```

````python
%matplotlib inline

import matplotlib.pyplot as plt
import modules as mod
````

* 1\.2\. Loome mudeli <strong> NB! (invariantset) skalaarvälja tähista muutujaga 'x' </strong>

Mudelil on võimalik defineerida ka nimi. See on kasulik, kui ühele graafikule hiljem kuvatakse mitu mudelit. Nime aga ei pea defineerima.

```Python
    mudel = mod.InflationModel(A="1", B="1", V="x**p", name="Mudel-1")
```

See loob mudeli, kus funktsioonideks on:
<center>1. $A = 1$</center>
<center>2. $B = 1$</center>
<center>3. $V = x^p$</center>

Pärast defineerimist küsib programm vabade parameetrite väärtusi. Arvutamisel koostatakse kõik võimalikud kombinatsioonid parameetrite vahel.

````python
mudel = mod.InflationModel(A="1", B="1", V="x**p-1", name="Mudel-1")
````

* 1\.3\. Arvutamiseks kutsu välja funktsioon .calculate()

```Python
    mudel.calculate()
```
Arvuatmise käigus küsitakse vajadusel kasutajalt sisendit. Selline olukord võib tekkida näiteks, kui programm leiab vahemikust mitu erinevat võimalust määrata skalaarvälja lõppväärtus.

Arvutamisel saab määrata ka, mis meetodil proovitakse esimesena lahendada. Vaikimisi lahendatakse numbriliselt.
Selleks tuleb vastavalt .calculate() sisse määrata meetod:

```Python
    mudel.calculate(method=#'numeric' või 'symbolic'#)
```
Samuti on võimalik kuvada, kas kasutaja soovib arvutamise protsessit rohkem infot:

```Python
    mudel.calculate(info=#True / False#)
```

````python
mudel.calculate(info=False)
````

## 2. Programmi kasutamine $I_V(I_\phi)$ korral

Sellel juhul on programmi kasutamine kõik sarnane nagu $A(\phi)$, $B(\phi)$ ja $V(\phi)$ korral. Erinevuseks on ainult mudeli defineerimine.

```Python
    mudel = InflationModel(IV="(1-exp(-sqrt(2/(3*a))*x))**2")
```

See kood loob mudeli, mille $I_V$ funktsiooniks on $I_V = \left[1-\exp{\left(-\sqrt{\frac{2}{3 \cdot a}}\right)} \cdot x)\right]^2$

````python
mudel2 = mod.InflationModel(IV="(1-exp(-sqrt(2/(3*a))*x))**2", name="Mudel-2")
````

````python
mudel2.calculate(info=False)
````

# Graafikud

Kui mudel on arvutused lõpetanud, siis saab kuvada graafikud. Selleks tuleb kutsuda funktsioon, et graafikud valmistada.

```Python
    mudel.plot_graph(plot_type=)
```

Graafikute kuvamiseks kutsuda välja funktsioon

```Python
    mudel.plot_show()
```

Võimalik on kuvada kolme tüüpi graafikuid. Vastavalt ka nende numbrid.
1. Skalaar spektraal indeksi - tensor-skalaarhäirituse suhte graafik.
2. Inflatioonimäära sõltuvus algsest skalaarvälja väärtusest
3. Slow-roll parameetri ja skalaarvälja väärtuse graafik

Variandina on veel võimalus määrata graafikud id number, mis võimaldab erinevate mudelite tulemusi koos kuvada. Selle puhul vaata ette, et erinevat tüüpi graafikud ei omaks samasid id numbreid. Vaikimisi on ka määratud väärtused.
Ning lisaks kas 1. graafiku puhul on huvi ka vastavate punktide kohta.

```Python
    mudel.plot_graph(plot_type=1, plot_id=101, info=True)
```

````python
mudel.plot_graph(plot_id=101, info=False)
mudel2.plot_graph(plot_id=101, info=False)
mudel.plot_show()
````
