# Mõjufunktsionaal

Inflatsiooni kirjeldamiseks tuuakse sisse uus füüsikaline väli - skalaarväli. Lihtsamal juhul on see toodud mõjufunktsionaali juurde liidetavana ja ei ole seotud gravitatsiooniga (Ricci skalaariga). Seda tüüpi teooriat nimetatakse minimaalselt seotud teooriaks. Sellisel juhul saab mõju funktsionaali kirja panna kujul

$$S = \frac{1}{2} \int d^4 x \sqrt{-g}\left( R - (\partial^\mu \phi)(\partial_\mu \phi) - \frac{V(\phi)}{2}\right) \ . \tag{1}$$
Selles mõjust saab R kaudu tuletada Einsteini võrrandid ning teise kahe liikme abil dünaamika võrrandid skalaarvälja jaoks. Tavalislt listakse mõjufunktsionaali veel mateeriat kirjeldav mõju liige, kuid antud juhul on tegemist inflatsiooni perioodiga ja sellel juhul on hea lähendus, et mateeriat ei ole veel jõudnud tekkida. Seega $S_{\text{mateeria}} = 0$.

Üldisemal juhul võib kirjeldada ka teooriat, kus skalaarväli on seotud Ricci skalaariga. Sel juhul nimetatakse teooria mitte-minimaalselt seotud teooriaks. Mõjufunktsionaali võib üldisemal juhul kirja panna kujul
$$S = \frac{1}{2} \int d^4 x \sqrt{-g}\left( A(\phi)R - B(\phi)(\partial^\mu \phi)(\partial_\mu \phi) - \frac{V(\phi)}{2}\right) \ . \tag{2}$$

Mõjufunktsionaalis on kirjeldatud kolm skalaarväljast sõltuvat funktsiooni: $A(\phi)$, $B(\phi)$ ja $V(\phi)$.
* $A(\phi)$ - funktsioon, mis määrab, kas skalaarväli on gravitatsiooniga seotud. Kui A(\phi) = 1, siis öeldakse, et tegemist on mitte-minimaalselt seotud teooriaga. Kuna gravitatsioon on tõmbav, siis rakendub A funktsioonile ka tingimus $0 < A(\phi) < \infty$.

* $B(\phi)$ - seosefunktsioon. Piiranguid ei ole.

* $V(\phi)$ - on varasemast tuttav skalaarvälja potentsiaali kirjeldav funktsioon. Üldiselt sellele funktsioonile väga rangeid piiranguid ei ole, kuid levinud on siiski positiivsete väärtustega potentsiaalid.


Lisaks saab tuletada veel ühe soese, et vältida kummitusvälju (*ghost field*), mis võivad tekitada teooriasse mitte füüsikalisi osakesi, nagu osakesed mille kineetiline energia on negatiivne. Seega peab olema ka täidetud tingimus
$$\frac{B(\phi)}{A(\phi)} + \frac{3}{2}\left(\frac{A'(\phi)}{A(\phi)} \right)^2 > 0 \tag{3}$$

## Teisendused ja invariandid

Selgub, et leiduvad teisendused (parametriseeringud), mis jätavad mõju (2) funktsionaalse kuju samaks. Esimene neist on teisendus skalaarväljale
$$\phi = \overline{f}(\overline{\phi}) \ , \text{ kus } \overline{f}' > 0 \tag{4}$$
ning teine kasulik funktsioon on konformne teisendus, mis jätab vektorite vahelised nurgad muutumatuks, aga muudab nende pikkusi
$$g_{\mu\nu} = \exp\Big(2\overline{\gamma}(\overline{\phi})\Big) \cdot \overline{g}_{\mu\nu} \tag{5}$$
(ülemise joonega on tähistatud suuruseid, mis on teises parametriseeringus.)


Mõned kasulikud omadused:
* Saab viia mõju kujule, kus arvutusi on kergem teha.
* Erinevad funktsioonide $A(\phi)$, $B(\phi)$ ning $V(\phi)$ kombinatsioonid annavad samad tulemused, kui arvutada vaatuslikult võrreldavad suurused.

Nende teisenduste abil on võimalik seega tuletada arvutuseeskirjad, mis viivad läbi teisendusi $A \rightarrow \overline{A}$, $B \rightarrow \overline{B}$, $V \rightarrow \overline{V}$. Neid teisendusi kutsutakse Flanagani teisendusteks. Programmi seisukohalt tuleb mainida, et leidub kaks erinevad eeskirja:
* meetrilise formalismi jaoks - seostus $\Gamma^{\mu}_{\nu \rho}$ sõltub meetrikast.

* Palatini formalismi jaoks - seostus $\Gamma^{\mu}_{\nu \rho}$ ei sõltu meetrikast ja mõju varieerimisel tuleb seda arvesse võtta.

Nende erinevuste tõttu tulevad arvutamisel erinevad tulemused. Erinevus selle programmi kontekstis on see, et arvutamisel tuleb meetrilise formalismi korral rohkem liikmeid. Edasit kasutatakse notatsiooni formalismide eristamiseks
\begin{equation}
\delta_\Gamma = 
\begin{cases}
   0 \ , & \text{ Palatini formalism}\\
   1 \ , & \text{ meetriline formalism}\\
\end{cases} \tag{6}
\end{equation}

Flanagani teisendustele on võimalik ka tuletada invariantseid suuruseid. Ehk need suurused jäävad teisendustel muutumatuks. Järgnevalt on välja toodud kaks olulisemat, mida kasutatakse programmis ära:


$$I_V(\phi) = \frac{V(\phi)}{A^2(\phi)}$$


$$I_\phi(\phi) = \int \sqrt{\frac{B(\phi)}{A(\phi)} + \delta_\Gamma \frac{3}{2} \left(\frac{A'(\phi)}{A(\phi)}\right)^2} d\phi = \int \sqrt{F(\phi)}d\phi \tag{7}$$

## Programmi töö

Programmis kasutatakse ära asjaolu, et leidub invariantseid suuruseid, mis teisendustel ei muutu. Nende abil on võimalik näidata, kas erinevad mudelid (loe $A(\phi)$, $B(\phi)$ ning $V(\phi)$ kombinatsioonid) on ekvivalentsed ning ennustavad samasid vaatluslikult ekvivalentseid tulemusi.

Tulemuste arvutamisel eeldatakse, et tegemist on aeglase veeremise lähendiga. Sellest tulenevalt on defineeritud kaks aeglase veeremise parameetrit:

$$\varepsilon(\phi) = \frac{1}{2} \left(\frac{1}{I_V} \frac{dI_V}{dI_\phi} \right)^2 \tag{8}$$
$$\eta(\phi) = \frac{1}{I_V} \frac{d^2I_V}{dI_\phi^2} \tag{9}$$
Neist $\varepsilon(\phi)$ kaudu määratakse inflatsiooni lõpp: $\varepsilon(\phi) = 1$.

Aegalse veeremise parameetrite kaudu ongi arvutatud kaks suurust, mis on võrreldavad vaatlustega:


* tensor-skalaar häirituste suhe
$$r = 16\varepsilon(\phi) \tag{10}$$ 


* skalaarsete häirituste spektraalne indeks
$$n_s = 1 - 6\varepsilon(\phi) + 2 \eta(\phi) \tag{11}$$

# Funktsiooni $I_V$ ning funktsioonide $A$, $B$, $V$ abil koostatud mudelite võrdlemine.

Antud funktsionaaluse kasutamiseks peab kasutaja määrama ära nii $A(\phi)$, $B(\phi)$, $V(\phi)$ ja $I_V(I_\phi)$ funktsioonid.

Selles meetodis kasutatakse kahte arvutusskeemi, et tuletatada võrreldavad suurused. 

1) $dI_{\phi1}$:

Selle suuruse leidmiseks võetakse suuruset $I_\phi(\phi)$ tuletis ning saadakse tulemuseks $\sqrt{\frac{B}{A} + \delta_\Gamma \frac{3}{2} \left(\frac{A'}{A}\right)^2}$

2) $dI_{\phi2}$:

Arvutamine toimub mitmes etapis.
    a) Esimesena leitakse $I_V(I_\phi)$ pöördfunktsioon(id). Kui leidub mitu pöördfunktsiooni, siis kuvatakse tulemus iga pöördfunktsiooni jaoks, et oleks näha hiljem tulemustes sobivad määramispiirkonnad.
    b) Teisena asendatakse $I_\phi(I_V)$ funktsiooni suurus $\frac{V}{A^2}$, et tuletada funktsiooni $I_\phi$ sõltuvust $\phi$-st.
    c) Võetakse saadud suurusest tuletis.

Antud meetodil on probleemseks asjaolu, et võrreldakse tuletisi, mitte funktsioone endid. Põhjus sellele lähenemisele on sama, mis hiljem tekitab probleeme. Kui võrdlust teha nii, et võrrelda näiteks invariantset potentsiaali $\frac{V}{A^2}$ ja  asendades $I_V(I_\phi)$ funktsioonis $I_\phi$ vastava integraaliga, siis tekib probleem integreerimisel (lisandub konstant ning kõiki funktsioone ei saa arvutiga integreerida). Seepärast on alternatiivseks lahenduseks tuletise võtmine, mis näitab, et $A(\phi)$, $B(\phi)$ ja $V(\phi)$ ning $I_V(I_phi)$ kaudu arvutatud suurused on $I_\phi$ funktsiooni konstanti täpsusega.

Meetodi kasutamiseks kutsuda välja

```Python
    mudel.check_models(mudeli_parameetrid)
```

Kus mudeli_parameetrite asemel tuleb defineerida kõik vabad parameetrid. Näiteks kui leidub ainult kaks vaba parameetrit "a" ja "b" siis:

```Python
    mudel.check_models(a=1, b=5)
```

Tulemused kuvatakse kahel graafikul. Esimsel graafikul on kuvatud absoluutväärtus erinevuse suhtest $\frac{dI_{\phi1} - dI_{\phi2}}{dI)_{\phi1}}$ sõltuvuses $\phi$ väärtusest. Y-teljel on tulemused logaritmilises skaalas.
Teisel graafikul on kuvatud $dI_{\phi1}$ ja $dI_{\phi2}$ väärtused.

Iga pöördfunktsiooni jaoks on tulemused kuvatud eraldi ridadel.


````python
mudel3 = mod.InflationModel(A="2/(3*a)*x^2", B="1-4/a", V="4/(9*a^2)*(1-1/x)^2*x^4", IV="(1-exp(-sqrt(2/(3*a))*x))**2", name="Mudel-3")
````

````python
mudel3.check_models(a=1)
````

Kuvatud näite puhul on näha, et määramispiirkonnas $(0, 1)$ on sobiv pöördfunktsioon teine lahend. vahemikus (1, 10+] on sobiv pöördfunktsioon lahend 1 kuna viga on suurusjärgus $10^{-14}$ mis on peamiselt tingitud numbrilise arvutamise täpsusest.
